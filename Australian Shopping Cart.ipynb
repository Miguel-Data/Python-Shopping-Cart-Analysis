```python
#Univariate Data analysis
def univariate_analysis(data,color,title1,title2):
    fig, (ax1, ax2) = plt.subplots( #subplots
        ncols=2, #num of cols
        nrows=1, #num of rows
        figsize=(20,6) #set the width and height
    )
        
    
#Creating 2 plots below: Displot & ggplot
    sns.distplot( #Create a displot visualization
        data, #data
        ax=ax1, #axes 1 
        kde=True, #kde
        color = color #color
    )

    ax1.set_title(#Setting the title
        title1, 
        weight="bold",
        fontname="monospace",
        fontsize=25,
        pad=30
    )

    qqplot( #qqplot (quantile plot)
        data, #data
        ax=ax2, #axes2
        line='s' #line
    )

    ax2.set_title( #set the title 2
        title2,
        weight="bold",
        fontname="monospace",
        fontsize=25,
        pad=30
    )
    
    return fig #returning the figure
```
